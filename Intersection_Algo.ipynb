```python
import numpy as np
import pandas as pd
import descartes
import matplotlib.pyplot as plt
import geopandas as gpd
from geopandas import GeoDataFrame
from shapely.geometry import Point
%matplotlib inline
```

```python
df = pd.read_csv('bengaluru_latlng.csv')
data_len = df.shape[0]
status = np.random.randint(0,2,(data_len,1))
#status = np.zeros((data_len,1))
#status[50] = 1
#status[48] = 1
df['Status'] = status
d = df.values[:,2:4]
```

```python
geometry = [Point(xy) for xy in zip(df.Lat, df.Lon)]
df = df.drop(['Lat','Lon'], axis=1)
crs = {'init': 'epsg:4326'}
gdf = GeoDataFrame(df, crs=crs, geometry=geometry)
print(gdf)
```

```python
print(gdf.geometry[0])
```

```python
data_contact = []

for i in range(data_len):
    if (gdf.Status[i]==1):

        gdf.to_crs(epsg=3310,inplace=True)
        l = gdf.distance(gdf.geometry[i])
        l = np.array(l)
        data = gdf[l < 500].values
        for j in range(len(data)):
            if(data[j][1] != 1):
                data_contact.append(data[j])
                
data_contact = pd.DataFrame(data_contact,columns=['Addr','Status','Location'])
data_contact.drop_duplicates(subset ="Addr",keep = "first", inplace = True)
print(data_contact)
```
